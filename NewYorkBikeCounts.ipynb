```python
# Dependencies
import pandas as pd
import plotly.express as plt
import os
import statsmodels.api as sm
```

```python
# Import dataset from file
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])

# Remove the index column
bikecounts_df.drop(columns=['Unnamed: 0'], inplace=True)

# Calculate the median temperature
bikecounts_df['Median Temp (°F)']=(bikecounts_df['High Temp (°F)']+bikecounts_df['Low Temp (°F)'])/2

bikecounts_df.head(10)
```

#Hypothesis
There is a positive correlation between temperature and the number of bikes crossing the brooklyn bridge.

```python
independent_variables=bikecounts_df[['Median Temp (°F)']]
dependent_value=bikecounts_df['Brooklyn Bridge']
X=sm.add_constant(independent_variables)
regression_model=sm.OLS(dependent_value,X)
results=regression_model.fit()
```

```python
print(results.summary())
```

```python
results.summary2()
```

##Conclusion
The P-value for Median Temperature is 0.0001 we find significant evidence to reject the null hypothesis. There is a positive correlation between the number of bikes crossing the brooklyn bridge and the median temperature.
