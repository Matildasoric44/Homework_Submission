```python
# Import libraries and dependencies
import pandas as pd
import plotly.express as px
```

```python
# Read the CSV file into a `DataFrame` and set the date column as the index
df= pd.read_csv('../Resources/stocks.csv')
df.head()
```

```python
# Use the `pct_change` function in Pandas to calculate daily returns for each stock and drop N/A values
df["daily_returnJNJ"]=df['JNJ'].pct_change(1).dropna()
df["daily_returnHD"]=df['HD'].pct_change(1).dropna()
df["daily_returnINTC"]=df['INTC'].pct_change(1).dropna()
df["daily_returnAMD"]=df['AMD'].pct_change(1).dropna()
df["daily_returnMU"]=df['MU'].pct_change(1).dropna()
df["daily_returnNVDA"]=df['NVDA'].pct_change(1).dropna()
df["daily_returnTSM"]=df['TSM'].pct_change(1).dropna()
df=df.dropna()
print(df)
```

```python
# Use the `corr` function to calculate correlations of their daily returns for each stock pair
#daily_returnJNJ.corr(daily_returnJNJ,)
#daily_returnHD.corr()
#daily_returnINTC.corr()
#daily_returnAMD.corr()
#daily_returnMU.corr()
#daily_returnNVDA.corr()
#daily_returnTSM.corr()
corrdf=df.corr(method='pearson')
print(corrdf)
```

```python
# Create a heatmap from the correlation values
import seaborn as sns
sns.heatmap(corrdf, annot=True)
```
